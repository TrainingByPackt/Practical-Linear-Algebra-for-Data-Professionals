# Example 8.6.1

```python
import pandas as pd
data = pd.read_csv('movie_data.csv',index_col=0)
data.head()
```

```python
data = data.pivot(index='UserID',columns='MovieID',values='Rating').fillna(0)
data.head()	
```

## Example 8.6.2

```python
import numpy as np
from numpy import dot

def cosine_similarity(x,y):
    X_m = np.sqrt(x.dot(x.T)[0][0])
    Y_m = np.sqrt(y.dot(y.T)[0][0])
    dotproduct = x.dot(y.T)[0][0]
    return dotproduct/(X_m*Y_m)

user1 = np.array(data.loc[data.index==1])
user18 = np.array(data.loc[data.index==18])

cosine_similarity(user1,user18)
```

# Example 8.6.3

```python
from scipy.linalg import svd
import numpy as np
from numpy import zeros
from numpy import dot

U, s, VT = svd(data)

m,n = data.shape[0], data.shape[1]

sigma = np.zeros((m, n))
for i in range(min(m, n)):
    sigma[i, i] = s[i]
```

Specify the number of singular values to keep.

```python
#specify the number of singular values to keep
keep_values = 50
#select first few columns of U
U_c = U[:,:keep_values]
#select  first few columns and few rows of U
sigma_c = sigma[:keep_values,:keep_values]
#select first few rows of VT
VT_c = VT[:keep_values,:]


#now multiply
data_compressed = np.dot(U_c,np.dot(sigma_c,VT_c))

data_compressed = pd.DataFrame(data_compressed)
data_compressed.columns.name = 'MovieID'
data_compressed.index.name='UserID'
data_compressed.index = data_compressed.index + 1
print("Compressed, after approximating from the first "+str(keep_values)+" latent features")
data_compressed.head()
```

```python
user1_compress = np.array(data_compressed.loc[data_compressed.index==1])
user18_compress = np.array(data_compressed.loc[data_compressed.index==18])

cosine_similarity(user1_compress,user18_compress)
```

## Example 8.6.4

```python
doc0 = "The dog bit Johnny"
doc1 = "Johnny bit the dog"
doc2 = "Why did the chicken cross the road"
doc3 = "This sounds challenging"
doc4 = "The bird was smashed on the road"
doc5 = "A boy named Johnny has a dog"
doc6 = "A dog is a grown puppy"
doc7 = "The moon comes out at night"

sentences = [doc0,doc1,doc2,doc3,doc4,doc5,doc6,doc7]
string=' '.join(sentences)
unique_words = set(string.split(' '))

word_count = {}
for word in unique_words:
    doc_count = {}
    for doc_num in range(len(sentences)):
        doc_count[doc_num] = sentences[doc_num].count(word) 
    word_count[word]=doc_count

import pandas as pd
data = pd.DataFrame(word_count).transpose()
data.columns.name = 'sentence#'
data.index.name = 'words'
data
```
